```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('mashicgetda.jpg')
img = cv2.resize(img,dsize=(600, 400))
cv2_imshow(img)
```

```python
img = cv2.imread('mashicgetda.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img,dsize=(600, 400))
cv2_imshow(img)
```

```python
import numpy as np
# Sobel Filter
filter = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
```

```python
f_img = cv2.filter2D(img, -1, filter) # edge 추출(강조)
cv2_imshow(f_img)
```

```python
filter = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])
```

```python
f_img = cv2.filter2D(img, -1, filter)
cv2_imshow(f_img)
```

```python
filter = np.array([[1/9,1/9, 1/9],
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]])
```

```python
f_img = cv2.filter2D(img, -1, filter)
cv2_imshow(f_img)
```

```python
f_img = cv2.filter2D(img, -1, filter)
f_img = cv2.filter2D(f_img, -1, filter)
f_img = cv2.filter2D(f_img, -1, filter)
cv2_imshow(f_img)
```

```python
# Gaussian Filter
filter = np.array([[1/16,1/8, 1/16],
                   [1/8, 1/4, 1/8],
                   [1/16, 1/8, 1/16]])
```

```python
f_img = cv2.filter2D(img, -1, filter)
cv2_imshow(f_img)
```

```python
# canny filter
filter = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
```

```python
f_img = cv2.filter2D(img, -1, filter)
cv2_imshow(f_img)
```
